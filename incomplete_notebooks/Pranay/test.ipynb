```python
from equadratures import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
mybasis = Basis('total-order', [3,3])
```

```python
I = mybasis.get_elements()
```

```python
plt.plot(I[:,0], I[:,1], 'o', markersize=16)
plt.xlabel('$i_1$')
plt.ylabel('$i_2$')
plt.show()
```

```python
mybasis = Basis('hyperbolic-basis', [3,3], q=0.5)
I = mybasis.get_elements()
plt.plot(I[:,0], I[:,1], 'ro', markersize=16)
plt.xlabel('$i_1$')
plt.ylabel('$i_2$')
plt.show()
```
